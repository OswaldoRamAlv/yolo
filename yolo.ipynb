# ***Algoritmo yolo para deteccion de cascos en imagenes de partidos de Futbol Americano***

## *Objetivos del proyecto*

Este proyecto buscar utilizar el modelo yolo de ultralitics para la deteccion de objetos, con el fin de que pueda identificar los cascos de jugatores de Futbol Americano.

## *datos utilizados*

Los datos utilizados para este proyecto fueron obtenidos desde kaggle, el enlace de la competencia donde se alojan los datos se encuentra en el archvo README adjuntado en el repositorio de github de este proyecto.

Para este ejercicio no se utilizaron todos los datos, unicamente una muestra de las imagenes alojadas en la carpeta images, estas mismas se prepararon mediante la herramienta de anotaciones de roboflow.

## *Sobre el entorno utilizado*

Las dos librerias utilizadas en este ejercicio son roboflow y ultralitics,las cuales se alojaron desde un entorno de anaconda, se recomienda crear un entorno virtual en el que se incluyan estas dos librerias, de igual manera, todas las librerias utilizadas para este proyecto se encontraran en el archivo requirements.txt
